```python
import pandas as pd
steam=pd.read_csv('steam.csv')
steam
```

```python
steam_description_data=pd.read_csv('steam_description_data.csv')
steam_description_data
```

```python
steam_description_data.rename(columns={'steam_appid': 'appid'}, inplace=True)
steam_description_data
```

```python
steam_data=steam.merge(steam_description_data,on='appid')
```

```python
steam_data
```

```python
steam_data=steam_data.drop(columns=['release_date','english','required_age','achievements','average_playtime','median_playtime','owners','price','detailed_description','about_the_game','platforms','developer','publisher'])
```

```python
steam_data
```

```python
steam_data.loc[steam_data['negative_ratings'] == 0, 'negative_ratings'] = 1
```

```python
def convert(obj):
    return obj.split(';')
def convert2(obj2):
    return [obj2]
steam_data['categories']=steam_data['categories'].apply(convert)
steam_data['steamspy_tags']=steam_data['steamspy_tags'].apply(convert)
steam_data['genres']=steam_data['genres'].apply(convert2)
steam_data['short_description']=steam_data['short_description'].apply(convert2)
steam_data['new_column']=steam_data['categories']+steam_data['genres']+steam_data['steamspy_tags']+steam_data['short_description']
steam_data['ratings']=steam_data['positive_ratings']*100//(steam_data['positive_ratings']+steam_data['negative_ratings'])
steam_data['ratings']=steam_data['ratings'].apply(convert2)
steam_data['tag']=steam_data['ratings']+steam_data['new_column']
steam_data=steam_data.drop(columns=['genres','categories','steamspy_tags','positive_ratings','negative_ratings','short_description','new_column'])
```

```python
steam_data
```

```python
steam_data['tag'][0]
```

```python
steam_data['tag']=steam_data['tag'].apply(lambda x:[i.replace(" ","") for i in x if type(i)==str])
steam_data['tag']=steam_data['tag'].apply(lambda x:[i.replace("-","") for i in x if type(i)==str])
```

```python
steam_data['tag'][0]
```

```python
steam_data['tags']=steam_data['ratings']+steam_data['tag']
steam_data=steam_data.drop(columns=['ratings','tag'])
steam_data
```

```python
steam_data['tags'] = steam_data['tags'].apply(lambda x: " ".join(map(str, x)) if isinstance(x, list) else x)
```

```python
steam_data['tags'][0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
steam_data['tags']=steam_data['tags'].apply(lambda x: x.lower())
```

```python
steam_data['tags']=steam_data['tags'].apply(stem)
```

```python
steam_data['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features = 28000,stop_words='english')
vectors=cv.fit_transform(steam_data['tags']).toarray()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=28000, stop_words='english')
vectors = cv.fit_transform(steam_data['tags'])

# Compute cosine similarity directly without converting to dense array
similarities = cosine_similarity(vectors, dense_output=False)
```

```python
similarities[1]
```

```python
similarities_dense = similarities[1].toarray()
```

```python
similarities_dense
```

```python
dense_similarity_matrix = similarities.toarray()
```

```python
similarity=dense_similarity_matrix
```

```python
steam_data[steam_data['name']=='Cyber Gun'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x: x[1])[1:6]
```

```python
steam_data.iloc[0]['name']
```

```python
def recommend(game):
    idx=steam_data[steam_data['name']==game].index[0]
    distances=similarity[idx]
    games_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]
    for i in games_list:
        print(steam_data.loc[i[0]]['name'])
    return
```

```python
recommend("Hitman: Blood Money")
```
